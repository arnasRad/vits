```python
import numpy as np
import utils
from mel_processing import spectrogram_torch
from pathlib import Path
from utils import load_wav_to_torch
import matplotlib.pyplot as plt
from scipy import signal
from scipy.io import wavfile

filepath = Path('/home/arnas/Downloads/tawgvaw/MIF0000_0020_ar_kunigiska_jo_charakteri_jie_stiprino_ar_skalde.wav')

audio, sample_rate = load_wav_to_torch(filepath)
audio_norm = audio / 32768.0
audio_norm = audio_norm.unsqueeze(0)
spec = spectrogram_torch(audio_norm, 1024, 22050, 256, 1024, center=False)
fig, ax = plt.subplots(figsize=(10, 2))
im = ax.imshow(spec.data[0].cpu().numpy(), aspect="auto", origin="lower",
               interpolation='none')
plt.colorbar(im, ax=ax)
plt.xlabel("Frames")
plt.ylabel("Channels")
plt.tight_layout()
```
